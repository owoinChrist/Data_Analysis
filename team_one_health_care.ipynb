```python
# Checks for size of neighbourhood who show up
present = []
for a in neigh_noshow_No['Neighbourhood']:
    present.append(a)
print(len(present))
```

```python
# Checks for size of neighbourhood who did not come
absent = []
for a in neigh_noshow_Yes['Neighbourhood']:
    absent.append(a)
print(len(absent))
```

```python
# Checks for unique present neighbpurhood 
unique_present_name = []
for a in present:
    if a not in absent:
        unique_present_name.append(a)
    else:
        continue

print(unique_present_name)
```

```python
# Checks for unique absent neighbpurhood 
unique_absent_name = []
for a in absent:
    if a in present:
        continue 
    else:
        unique_absent_name.append(a)

print(unique_absent_name)
```

```python
# Gets wait time
waittime = (df1['AppointmentDay'] - df1['ScheduledDay']).dt.days
waittime
```

```python
# Creates waittime column
df1['waittime'] = waittime
```

```python
df1[['Age Group', 'waittime']].value_counts().reset_index()
```

```python
# Categorises waittime
bins = [0, 10, 40, 60, 80, 100, 120, 140, 160, 180]
labels = ['0-10', '11-40', '41-60', '61-80', '81-100', '101-120', '121-140', '141-160', '161-180'] 
waitInterval = pd.cut(df1.waittime, bins=bins, labels=labels,
      include_lowest=True)
waitInterval
```

```python
# Creates waittime interval column
df1['WaitInterval'] = waitInterval
```
